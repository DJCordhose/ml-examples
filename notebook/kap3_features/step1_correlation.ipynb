```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../datasets/Iris/iris_cleaned.csv')
```

```python
df.boxplot()
```

```python
df['class'].unique()
```

Encode class from categorical to numerical value for correlation plot

```python
df['class num'] = df.apply(lambda row: 0 if row['class'] == 'Iris-setosa' else (1 if row['class'] == 'Iris-versicolor' else 2),axis='columns')
```

```python
corrmat = df.corr()
sns.heatmap(corrmat, annot=True)
```

```python
sepal_length = df['sepal length']
petal_length = df['petal length']
sns.jointplot(sepal_length, petal_length)
```

```python
petal_width = df['petal width']
sns.jointplot(petal_width, petal_length)
```

```python
sepal_width = df['sepal width']
sns.jointplot(sepal_width, sepal_length)
```

```python
class_num = df['class num']
sns.jointplot(class_num, petal_width)
```

```python
sns.jointplot(class_num, sepal_width)
```
